```python
"""Ejercicio 1

Investigue y defina:
● ¿Qué es un cuaderno (notebook) Jupyter? ¿Que es una celda? ¿Que tipos de celdas existen?
● ¿Qué es un entorno de ejecución? ¿Cuánto tiempo dura una sesión de un entorno?



¿Qué es un cuaderno (notebook) Jupyter?
Un cuaderno Jupyter es una aplicación web interactiva que permite crear y compartir documentos que contienen código ejecutable, ecuaciones, visualizaciones y texto narrativo. Se utiliza principalmente para el análisis de datos, aprendizaje automático, simulaciones y otras tareas relacionadas con la programación. Los cuadernos Jupyter son populares porque permiten ejecutar código de manera secuencial y ver los resultados en tiempo real, lo que es útil para experimentar y documentar procesos.

¿Qué es una celda?
En un cuaderno Jupyter, una celda es una unidad de trabajo donde puedes escribir y ejecutar contenido. Las celdas permiten estructurar y organizar el trabajo en un cuaderno, separando diferentes tipos de contenido como código, texto o visualizaciones.

Tipos de celdas en un cuaderno Jupyter:
Celdas de código: Son celdas en las que se escribe código en un lenguaje de programación, como Python. Cuando se ejecuta una celda de código, el contenido se envía al kernel para ser procesado, y los resultados (salida del código) se muestran justo debajo de la celda.

Celdas de texto (Markdown): Estas celdas permiten escribir texto en formato Markdown, lo que incluye títulos, listas, enlaces, imágenes, y más. Son útiles para agregar descripciones, explicaciones, y documentación en el cuaderno.

Celdas de salida: Estas celdas muestran los resultados o la salida generada por la ejecución de las celdas de código, como gráficos, tablas o texto.

¿Qué es un entorno de ejecución?
Un entorno de ejecución es el entorno computacional donde se ejecuta el código de un cuaderno Jupyter. Incluye el kernel que interpreta y ejecuta el código, así como los recursos de hardware (CPU, RAM, etc.) asignados para esa ejecución. El entorno de ejecución puede estar en tu máquina local, en un servidor remoto o en la nube.

¿Cuánto tiempo dura una sesión de un entorno?
La duración de una sesión de un entorno de ejecución depende de varios factores:

Configuración del servidor: Si estás usando Jupyter en un servidor remoto o en la nube, la duración de la sesión puede estar limitada por la configuración del servidor o los recursos disponibles.

Inactividad: Algunos entornos se configuran para cerrarse automáticamente después de un período de inactividad.

Uso de recursos: Si se consumen demasiados recursos, el entorno puede cerrarse para liberar memoria o CPU.

En general, una sesión puede durar desde unos minutos hasta horas, dependiendo de las políticas del entorno en el que estés trabajando.

"""
```
